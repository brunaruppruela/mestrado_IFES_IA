```python
from transformers import pipeline
import os
```

```python
# Criando o pipeline para análise de sentimentos
pipe = pipeline("sentiment-analysis")

# Exemplo de uso do pipeline para analisar uma frase positiva
positive_sentence = "This movie is fantastic, I loved every moment of it!"
result_positive = pipe(positive_sentence)
print("Sentiment analysis for positive sentence:", result_positive)

# Exemplo de uso do pipeline para analisar uma frase negativa
negative_sentence = "I couldn't stand this movie, it was terrible."
result_negative = pipe(negative_sentence)
print("Sentiment analysis for negative sentence:", result_negative)
```

```python
# Criando o pipeline para análise de sentimentos
pipe = pipeline("sentiment-analysis")

# Caminho para o diretório onde os dados estão armazenados
data_dir = "./aclImdb/train/"

# Lista para armazenar as frases e seus sentimentos
sentences_sentiments = []

# Percorrendo os arquivos e pegando as 10 primeiras frases
count = 0
for file_name in os.listdir(data_dir + "pos/")[:10]:
    with open(data_dir + "pos/" + file_name, 'r', encoding='utf-8') as file:
        sentence = file.read()
        # Aplicando o pipeline para análise de sentimento
        result = pipe(sentence)[0]
        sentences_sentiments.append((sentence, result))
        count += 1
        if count == 100:
            break

for file_name in os.listdir(data_dir + "neg/")[:10]:
    with open(data_dir + "neg/" + file_name, 'r', encoding='utf-8') as file:
        sentence = file.read()
        # Aplicando o pipeline para análise de sentimento
        result = pipe(sentence)[0]
        sentences_sentiments.append((sentence, result))
        count += 1
        if count == 200:
            break

# Exibindo os resultados
for idx, (sentence, sentiment) in enumerate(sentences_sentiments, start=1):
    print(f"Sentence {idx}:")
    print("Text:", sentence)
    print("Sentiment:", sentiment)
    print()
```
